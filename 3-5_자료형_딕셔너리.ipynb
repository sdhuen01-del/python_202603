## 5.딕셔너리 관련 함수

### keys() : Key 리스트 만들기

```python
dic = {'name': '안정운', 'phone':'010-1111-2222', 'birth': '1111'}
dic.keys()
```

```python
list(dic.keys())
```

```python
for key in dic.keys():
    print(key)
```

### values() : Value 리스트 만들기

```python
dic.values()
```

```python
for value in dic.values():
    print(value)
```

### items() : key, value 쌍 얻기

```python
dic.items()
```

### clear() : key, value 쌍 모두 지우기

```python
dic
```

```python
dic.clear()
```

```python
dic
```

### get() : Key로 value 얻기

```python
a = {'name': '안정운', 'phone': '010-1111-2222', 'birth': '1111'}
```

```python
a['name']
```

```python
a.get('name')
```

```python
print(a.get('age'))
```

```python
a['age']
```

```python
a
```

```python
a.get('age', 20)
```

```python
a['age'] = 25
a
```

```python
a.get('age', 20)
```
