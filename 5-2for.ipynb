## 3.for문의 응용

```python
marks = [90,25,67,45,80] # 학생들 시험 점수
number = 0

for mark in marks:
    number += 1 # number = number +1
    if mark >=60:
        print('%d번 학생은 합격입니다.'%number)
    else:
        print('%d번 학생은 불합격입니다.'%number)
```

##4.for문과 continue문

```python
marks
```

```python
number
```

```python
number = 0

for mark in marks:
    number += 1
    if mark < 60:
        continue
    print('%d번 학생은 합격입니다.'%number)
```

## 7. for와 range를 이용한 구구단

```python
for i in range(2, 10): # 구구단의 단
    for j in range(1, 10):
        print(i*j, end=' ')
    print()
```

## 8.리스트 컴프리헨션 사용하기

```python
a = [1,2,3,4]
result=[]

for num in a:
    result.append(num * 3)

print(result)
```

```python
# 리스트 컴프리헨션(list comprehension)
a = [1,2,3,4]
result = [num*3 for num in a]
print(result)
```

```python
#짝수만 3을 곱하여 result에 리스트 컴프리헨션으로 표현
a = [1,2,3,4]
result = [num*3 for num in a if num%2 == 0]
print(result)
```

```python
a = [1,2,3,4]
result=[]

for num in a:
    if num%2 == 0 :
        result.append(num * 3)

print(result)
```

```python
# 리스트 컴프리헨션으로 중첩 for문(구구단)
gogodan = [i*j for i in range(2,10)
              for j in range(1,10)]
print(gogodan)
```
