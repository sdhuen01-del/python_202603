## 1. 계산기 프로그램을 만들며 클래스 알아보기

```python
result = 0 #초기 설정값

def add(num):
    global result
    result += num
    return result

print (add(3))
print (add(4))
```

```python
# 2대의 계산기
result1 = 0
result2 = 0

def add1(num):
    global result1
    result1 += num
    return result1

def add2(num):
    global result2
    result2 += num
    return result2

print(add1(3))
print(add1(4))
print(add2(3))
print(add2(7))
```

```python
# 여러대 계산기를 클래스로 해결하기
class Calculator: # 클래스는 설계도
    def __init__(self): #언더바 2개는 파이썬 특별한 언어
        self.result = 0 # 초기값 설정

    def add(self, num):
        self.result += num
        return self.result
```

```python
cal1 = Calculator()
cal2 = Calculator()
```

```python
print(cal1.add(3))
```

```python
print(cal1.add(4))
```

```python
print(cal2.add(3))
```

```python
print(cal2.add(7))
```

## 4.생성자 (Constructor)

- 생성자는 객체가 생성될 때 자동으로 실행 되는 메서드
- 객체에 초기값을 설정하는 역할

```python
class FourCal: #클래스 선언
    def __init__(self, first, second): #생성자 함수 
        self.first = first
        self.second = second

    def add(self):
        result = self.first + self.second
        return result

    def sub(self):
        result = self.first - self.second
        return result

    def mul(self):
        result = self.first * self.second
        return result

    def div(self):
        result = self.first / self.second
        return result
```

```python
a = FourCal() #괄호 안에 값을 안 넣어서 오류남.
```

```python
a = FourCal(4,2)
```

```python
a.add()
```

```python
a.sub()
```

```python
a.mul()
```

```python
a.div()
```

## 6. 메서드 오버라이딩

```python
# 0으로 나눌 때 오류가 아닌 0을 리턴 하도록 처리 하려면?
class SafeFourCal(FourCal):
    def div(self): #메서드 오버라이딩 : 상속받는 클래스의 메서드 이름을 똑같이 사용하여 메서드를 재정의
        if self.second == 0:
            return 0
        else:
            return self.first / self.second
```
