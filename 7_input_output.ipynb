## 파일 읽고 쓰기
### 1) 파일 생성하기 : open()

```python
f = open('newfile.txt', 'w') #open close set로 다님 상대경로
f.close() #아무런 출력은 하지 않지만 폴더안에 파일이 생성. 파일 객체를 닫아 주는 역활을 함.
```

```python
# 특정 경로에 파일 생성
f = open('d:/newfile.txt', 'w') # 절대 경로
f.close() #추가 생성은 안되고 없으면 생성. 추가 실행 하면 파일 온오프만 됨
```

### 2) 파일을 쓰기 모드로 열어 내용 쓰기

```python
f = open('newfile.txt', 'w')

for i in range(1,11):
    data = '%d번째 줄입니다.\n'%i
    f.write(data) # 파일에 write

f.close()
```

```python
for i in range(1,11):
    data = '%d번째 줄입니다.\n'%i
    print(data)
```

### 3) 파일을 읽는 여러가지 방법

- readline() 함수 이용하기

```python
f = open('newfile.txt', 'r')
line = f.readline()
print(line)
f.close()
```

- readline()함수를 이용해서 여러줄 읽기

```python
f = open('newfile.txt', 'r')

while True:
    line = f.readline()
    if not line:
        break

    print(line)
f.close()
```

- 사용자의 입력을 받아 여러줄 출력

```python
while True:
    data = input()
    if not data :
        break
    print(data)
```

- readline() 함수 사용하기

```python
f = open('newfile.txt', 'r')
lines = f.readlines()
lines
```

```python
for line in lines:
    print(line)

f.close()
```

- 줄바꿈(\n)문자 제거 : strip()

```python
f = open('newfile.txt', 'r')
lines = f.readlines()

for line in lines:
    line = line.strip() # 줄바꿈 문자 제거
    print(line)

f.close()
```

- raed() 함수 사용하기 : 파일 내용 전체를 문자열로 반환

```python
f = open('newfile.txt', 'r')
data = f.read()
data
```

```python
print(data)
```

- 파일 객체를 for문으로 읽어오기

```python
f = open('newfile.txt', 'r')
for line in f:
    print(line)

f.close()
```

### 4) 파일에 새로운 내용 추가하기

```python
f = open('newfile.txt', 'a')

for i in range(11,21):
    data = '%d번째 줄입니다.\n'%i
    f.write(data)

f.close()
```

### 5) with문과 함께 사용하기 : 파일 열고 닫는것을 자동으로 처리

```python
with open('foo.txt','w') as f:
    f.write('Life is too short, you need python')
```
